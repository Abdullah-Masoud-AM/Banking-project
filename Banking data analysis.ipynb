```python
import pandas as pd
import numpy as np
```

```python
cards = pd.read_csv('C:\\Users\\Abdul\\Downloads\\Banking project\\cards.csv')
```

```python
users = pd.read_csv('C:\\Users\\Abdul\\Downloads\\Banking project\\users.csv')
```

```python
transactions = pd.read_csv('C:\\Users\\Abdul\\Downloads\\Banking project\\transactions.csv')
```

```python
cards['Credit Limit'] = cards['Credit Limit'].str.replace('$','',regex=False).astype(int)
```

### Cards analysis

#### Which card type is most commonly used?


```python
res = cards['Card Type'].value_counts().reset_index(name = 'count')
res.index = res.index + 1
res['count'] = res['count'].apply(lambda x : f"{x:,.0f}")
res
```

#### Which card brand is most commonly used?


```python
res = cards['Card Brand'].value_counts().reset_index(name = 'count')
res.index = res.index + 1
res['count'] = res['count'].apply(lambda x : f"{x:,.0f}")
res
```

#### How many cards in dark web?


```python
cards[cards['Card on Dark Web'] == 'yes'] 
```

#### what's the total card limit by card type? 

```python
res = cards.groupby('Card Type')['Credit Limit'].sum().reset_index(name = 'Total')
res.index = res.index + 1 
res['Total'] = res['Total'].apply(lambda x : f"{x:,.0f} $")
res
```

#### what's the total card limit by card brand? 

```python
res = cards.groupby('Card Brand')['Credit Limit'].sum().reset_index(name='Total')

res = res.sort_values(by='Total', ascending=False).reset_index(drop=True)

res.index = res.index + 1

res['Total'] = res['Total'].apply(lambda x: f"{x:,.0f} $")

res
```

### Transactions analysis

#### Number of transactions by year

```python
res = transactions['Year'].value_counts().reset_index(name = 'count').sort_values(by = 'count',ascending=False)
res.index = res.index + 1
res['count'] = res['count'].apply(lambda x : f"{x:,.0f}")
res
```

#### Number  of fraud transaction

```python
transactions[transactions['Is Fraud?'] == 'yes'].shape[0]
```

#### Number of transactions by chip

```python
res = transactions['Use Chip'].value_counts().reset_index(name = 'count').sort_values(by = 'count',ascending=False)
res.index = res.index + 1
res['count'] = res['count'].apply(lambda x : f"{x:,.0f}")
res 
```

#### Total amount per merchant city

```python
res = transactions.groupby('Merchant City')['Amount'].sum().reset_index(name = 'Total')
res = res.sort_values(by = 'Total',ascending=False).reset_index(drop=True)
res.index = res.index + 1
res['Total'] = res['Total'].apply(lambda x : f"{x:,.0f} $")
res
```

#### What's the total amount per merchant state?

```python
res = transactions.groupby('Merchant State')['Amount'].sum().reset_index(name = 'Total')
res = res.sort_values(by = 'Total',ascending=False).reset_index(drop=True)
res.index = res.index + 1
res['Total'] = res['Total'].apply(lambda x : f"{x:,.0f} $")
res
```

#### What's the total amount per year?

```python
res = transactions.groupby('Year')['Amount'].sum().reset_index(name='Total')

res = res.sort_values(by='Total', ascending=False).reset_index(drop=True)

res.index = res.index + 1

res['Total'] = res['Total'].apply(lambda x: f"{x:,.0f} $")

res
```

#### Number of transactions by errors

```python
res = transactions['Errors?'].value_counts().reset_index(name = 'count')
res = res.sort_values(by = 'count',ascending=False).reset_index(drop=True)
res.index = res.index + 1   
res
```
